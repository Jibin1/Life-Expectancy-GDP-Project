**1. Import all necessary libraries for the project**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**2. Assign a dataframe "df" that reads all the data stored in "all_data.csv"** <br>Print out the top few rows of the dataframe to see what columns and data it consists. 

```python
df = pd.read_csv('all_data.csv')
print(df.head())
```

The df consists of four columns: **Country, Year, Life expectancy at birth (years) and GDP**.<br> 
Country: Nominal Categorical Data<br>
Year: Ordinal Categorical Data<br>
Life expectancy: Continuous Numberical Data<br> 
GDP: Continuous Numerical Data<br>

**3. Print out the unique values of the categorical data: Country and Year**

```python
print(df.Country.unique())
print(df.Year.unique())
```

### Objective 1: Has GDP increased over time in the six nations?
**4. Create separate lists containing the GDP of each country.**<br>
\* Using a list is recommended as indexing through a dataframe will be deprecated in a future pandas update.

```python
chile_gdp = df.GDP[df.Country == 'Chile'].to_numpy()
china_gdp = df.GDP[df.Country == 'China'].to_numpy()
germany_gdp = df.GDP[df.Country == 'Germany'].to_numpy()
mexico_gdp = df.GDP[df.Country == 'Mexico'].to_numpy()
usa_gdp = df.GDP[df.Country == 'United States of America'].to_numpy()
zimbabwe_gdp = df.GDP[df.Country == 'Zimbabwe'].to_numpy() 
years_list = df.Year.unique()
country_list = ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe']
```

**5. Plot each country GDP from 2000 to 2015 using plt.plot**

```python
fig = plt.figure(figsize = (8,8))
plt.plot(years_list, chile_gdp, color = 'blue')
plt.plot(years_list, china_gdp, color = 'red')
plt.plot(years_list, germany_gdp, color = 'green')
plt.plot(years_list, mexico_gdp, color = 'black')
plt.plot(years_list, usa_gdp, color = 'orange')
plt.plot(years_list, zimbabwe_gdp, color = 'brown')
plt.legend(country_list)
plt.xlabel('Years (2000-2015)')
plt.ylabel('GDP')
plt.title('GDP From 2000-2015')
plt.show()
```

All six of the countries saw an increase in GDP during 2000-2015. However, the $\Delta$ GDP of some countries like Zimbabwe, and Chile was negligible when compared when compared to China and USA.  
**6. Quantify $\Delta$ GDP by plotting the difference between max and min GDP values of each country** 

```python
difference_list = [(chile_gdp[-1] - chile_gdp[0]), (china_gdp[-1] - china_gdp[0]), (germany_gdp[-1] - germany_gdp[0]),
                   (mexico_gdp[-1] - mexico_gdp[0]), (usa_gdp[-1] - usa_gdp[0]), (zimbabwe_gdp[-1] - zimbabwe_gdp[0])]
fig2 = plt.figure(figsize = (8,8))
plt.bar(country_list,difference_list)
plt.xlabel('Countries')
plt.ylabel('GDP')
plt.title('GDP Growth in 15 Years')
plt.show()
```
